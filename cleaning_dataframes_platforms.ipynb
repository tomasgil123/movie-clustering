```python
# ===> Vamos a filtrar las filas del dataframe que no tienen generos asociados

def filter_movies_with_out_genres(dataframe):
    
    array_len_strings = []

    for item in dataframe['genres']:
    
        array_len_strings.append(len(str(item)))
        
    dataframe['genre_empty'] = array_len_strings
    
    dataframe = dataframe[dataframe['genre_empty']>4]
    
    return dataframe
        
```

```python
# ===> Vamos a filtrar las filas del dataframe que son repetidas

# https://stackoverflow.com/questions/23667369/drop-all-duplicate-rows-in-python-pandas

def remove_duplicates(dataframe):
    
    dataframe = dataframe.drop_duplicates(subset=['id'])
    
    return dataframe
```

```python
df = filter_movies_with_out_genres(df)
```

```python
df = remove_duplicates(df)
```
